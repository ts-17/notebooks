```python
import numpy as np
import matplotlib.pyplot as plt

def appx_cantor_function(x, n_iter=20):
    a, b = 0.0, 1.0          # interval
    f_a, f_b = 0.0, 1.0      # vals at interval endpoints
    for _ in range(n_iter):
        third = (b - a) / 3.0
        left  = a + third
        right = b - third
        f_mid = 0.5 * (f_a + f_b)

        # x lies in left 1/3
        if x <= left:
            b, f_b = left, f_mid
        # x lies in right 1/3
        elif x >= right:
            a, f_a = right, f_mid
        # x lies in the removed middle 1/3
        else:
            return f_mid
    # linear intepr on the final (small) interval
    return f_a + (f_b - f_a) * (x - a) / (b - a)

cantor_vec = np.vectorize(appx_cantor_function)

# generate points
xs = np.linspace(0, 1, 4001)
ys = cantor_vec(xs, n_iter=25)

# plotting
plt.figure(figsize=(8, 4))
plt.plot(xs, ys)
plt.title("Cantor–Lebesgue Function")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.grid(True, linewidth=0.3)
plt.show()
```
